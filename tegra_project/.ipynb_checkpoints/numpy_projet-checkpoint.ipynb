# la forme et remodelage des tableaux numpy

```python
#nombre axes
tableau = np.array([[1,2,7],[3,4,8]])
print("array",tableau)
print("dimension",tableau.ndim)
```

```python
# pour recuperé la forme du tableau numpy on utilise la methode shape
c= np.shape([[1,2,2,5],[3,4,5,6],[5,5,6,8],[7,8,9,8],[7,8,9,8]])
print(c)
#pour recuperé juste le nombre de lignes ou le nombre de colonnes
c[0]#pour les lignes
```

```python
c[1]#pour les colonnes
```

```python
#la fonction size permet d'afficher la taille du tableau
d= np.size([[1,2,2,5],[3,4,5,6],[5,5,6,8],[7,8,9,8],[7,8,9,8]])
print(d)
```

```python
c[0]*c[1]#affichage de la taille du tableau de façon manuelle
```

```python
# la fonction reshape permet de modifier la forme du tableau (une ligne , le changer en ligne/colonne)
e = np.array([1,2,3,4,5,6])
e.reshape(3,-1)
e.reshape(-1,3)
```
